#### charger le dataset


```python
import pandas as pd

import os

BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath("__file__")))
file_path = os.path.join(BASE_DIR, "data", "customer_churn_business_dataset.csv")
# Charger les données
df = pd.read_csv(file_path)

# Voir les premières lignes
df.head()
```
